```python
sns.pairplot(df,hue='Clicked on Ad')
```

```python
plt.figure(figsize=(10, 7))
sns.scatterplot(x='Daily Time Spent on Site', y='Age', data=df,hue='Clicked on Ad' )
#It can be infer from this that younger people spent more time on the website and click the Ad as compare to elder people 
# It can be also seen that more the person spent time on site less are the chances that he/she will click on ad
```

```python
plt.figure(figsize=(10, 7))
sns.scatterplot(y='Daily Time Spent on Site', x='Area Income', data=df,hue='Clicked on Ad' )
# It can be infer from this plot that the richer section spent more time on site but they don't click on Ad on the other hand 
# area with less income click more on the Ad although they spent less time.
```

```python
from sklearn.model_selection import train_test_split
```

```python
df.drop(['Ad Topic Line', 'City', 'Country','Timestamp'], axis=1, inplace =True)
```

```python
X=df.drop('Clicked on Ad', axis=1)
y=df['Clicked on Ad']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
pred=lr.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test,pred))
```

```python
print(confusion_matrix(y_test,pred))
```
