```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./Data/data.csv")
```

```python
df.head()
```

```python
print("Columns in the dataset: ")
for i in df.columns:
    print(f"{i}", end=", ")
```

```python
#Number of left and right foot players
plt.figure(figsize=(18,8))
sns.countplot(x=df['Preferred Foot'],palette="Set2")
```

```python
#Position of players with preferred foot
plt.figure(figsize=(18,8))
# sns.set_theme(style="darkgrid")
sns.countplot(x=df.Position, hue=df['Preferred Foot'],palette="Set2")
```

```python
# shows that the range of players is from 15-45
# Most players has age in range of 20-25
plt.figure(figsize=(5,8))
sns.violinplot(y=df['Age'], color="gray")
```

```python
# Thi shows that mostly players who are younger have high potential than elder players
plt.figure(figsize=(18,8))
sns.lineplot(y=df['Age'], x = df['Potential'],color="green")
```

```python
plt.figure(figsize=(18,8))
sns.violinplot(x='Age', y='Potential', hue = 'Preferred Foot', palette='Set2', data =df, split=True )
```

```python
Age_10_to_20 = df.Age[(df.Age >= 10) & (df.Age <=20)]
Age_21_to_30 = df.Age[(df.Age >= 21) & (df.Age <=30)]
Age_31_to_40 = df.Age[(df.Age >= 31) & (df.Age <=40)]
Age_41_to_50 = df.Age[(df.Age >= 40) & (df.Age <=50)]

ageX = ["10-20", "21-30", "31-40", "41-50"]
ageY = [len(Age_10_to_20.values),len(Age_21_to_30.values),len(Age_31_to_40.values),len(Age_41_to_50.values)]
```

```python
plt.figure(figsize=(18,8))
sns.barplot(x=ageX, y=ageY, palette="mako")
```
